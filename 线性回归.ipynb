```python
import keras
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,100,300);
y = 3*x + 40 + np.random.randn(300)*5
```

```python
plt.scatter(x,y)
```

```python
model = keras.Sequential() #顺序模型
```

```python
from keras import layers #引入层
```

```python
model.add(layers.Dense(1,input_dim = 1))
```

```python
model.summary()
```

```python
#编译模型
model.compile(optimizer = 'adam',loss = 'mse')
```

```python
#训练模型
model.fit(x,y,epochs = 3000)
```

```python
model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),c = "r")
```
